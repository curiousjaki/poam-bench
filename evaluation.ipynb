```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
folder = "results/debug/"
```

```python
final_step_df = pd.read_csv(f'{folder}final_step_footprinting.csv',sep=";")
chained_df = pd.read_csv(f'{folder}activity_footprinting_chained.csv',sep=";")
unchained_df = pd.read_csv(f'{folder}proof_verification.csv',sep=";")
```

```python
composition_aggregated_secods = (pd.read_csv(f'{folder}activity_footprinting_unchained.csv',sep=";") + pd.read_csv(f'{folder}composition_footprinting_unchained.csv',sep=";"))["seconds"]
unchained_df["seconds"] = composition_aggregated_secods
```

```python
chained_df.columns = ["chained_"+ c for c in chained_df.columns]
unchained_df.columns = ["unchained_"+ c for c in unchained_df.columns]
final_step_df.columns = ["final_step_"+ c for c in final_step_df.columns]
```

```python
df_seconds = pd.concat([final_step_df, chained_df, unchained_df], axis=1)[["chained_seconds","unchained_seconds","final_step_seconds"]]
df_seconds.columns = ["Chained","Unchained","Final Step"]
df_seconds.plot(kind='line', title="Average Case Duration in Seconds", ylabel="Seconds", xlabel="Number of Activities", grid=True, logy=True)
```

```python
df_proof_size = pd.concat([final_step_df, chained_df, unchained_df], axis=1)[["chained_proof_size","unchained_proof_size","final_step_proof_size"]]
df_proof_size.columns = ["Chained","Unchained","Final Step"]
df_proof_size.plot(kind='line', title="Proof Size")
```

```python
# Create subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

# Plot on first subplot
# Plot on second subplot
df_proof_size.plot(kind='line', ax=ax[0], title="Proof Size", ylabel="Bits", xlabel="Number of Activities", grid=True, logy=True)
df_seconds.plot(kind='line', ax=ax[1], title="Instance Duration in Seconds", ylabel="Seconds", xlabel="Number of Activities", grid=True, logy=True)

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
